```python
def get_str_request(address):
    address = '+'.join(address.split())
    return 'https://search-maps.yandex.ru/v1/?apikey=' + apikey + '&text=' + address + '&type=biz&lang=ru_RU'
```

```python
def do_request(https):
    return json.loads(requests.get(https).text)
```

```python
def get_hours(data):
    n = 0
    while n < len(data['features']):
        if data['features'][n]['properties']['CompanyMetaData']['name'] == 'Пятёрочка':
            return data['features'][n]['properties']['CompanyMetaData']['Hours']['text']
        n = n + 1
    return (0)
```

```python

def get_hours_list(df):
    hl = []
    for i in range(len(df)):
        hl.append(get_hours(do_request(get_str_request(df['ADDRESS'][i]))))
    return hl
```

```python
hours_list = get_hours_list(df)
```

```python
with open('listfile.txt', 'w') as filehandle:
    for listitem in hours_list:
        filehandle.write('%s\n' % hours_list)
```

```python
def translete_hours_to_abs(hours_list):
    ahl = []
    for i in range(len(hours_list)):
        if hours_list[i] == 'ежедневно, 0:00–23:30':
            ahl.append(23.5)
        elif hours_list[i] == 'ежедневно, 7:00–22:00':
            ahl.append(15)
        elif hours_list[i] == 'ежедневно, 8:00–22:00':
            ahl.append(14)
        elif hours_list[i] == 'ежедневно, 8:00–23:00':
            ahl.append(15)
        elif hours_list[i] == 'ежедневно, 9:00–22:00':
            ahl.append(13)
        elif hours_list[i] == 'ежедневно, круглосуточно':
            ahl.append(24)
        else:
            ahl.append(np.nan)
    return ahl
        
```

```python
abs_hours_list = translete_hours_to_abs(hours_list)
```
